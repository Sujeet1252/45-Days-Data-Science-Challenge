```python
# ==============================
# SMART SCHOLARSHIP MANAGEMENT SYSTEM
# ==============================

print("===== STUDENT SCHOLARSHIP SYSTEM =====")

# Student Details
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

# Subject Marks
math = float(input("Math Marks: "))
science = float(input("Science Marks: "))
english = float(input("English Marks: "))
computer = float(input("Computer Marks: "))
hindi = float(input("Hindi Marks: "))

# Total & Percentage
total = math + science + english + computer + hindi
percentage = total / 5

# Grade Calculation
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
else:
    grade = "D"

# Pass / Fail
if (
    math < 33
    or science < 33
    or english < 33
    or computer < 33
    or hindi < 33
):
    status = "FAIL"
else:
    status = "PASS"

# Scholarship
if percentage >= 90:
    scholarship_percent = 100
elif percentage >= 80:
    scholarship_percent = 50
elif percentage >= 70:
    scholarship_percent = 25
else:
    scholarship_percent = 0

# Scholarship Amount
course_fee = 50000
scholarship_amount = (course_fee * scholarship_percent) / 100

# Result Display
print("\n===== STUDENT REPORT =====")

print(f"Name       : {name}")
print(f"Roll No    : {roll_no}")

print(f"\nTotal Marks : {total}")
print(f"Percentage  : {percentage:.2f}%")

print(f"Grade       : {grade}")
print(f"Status      : {status}")

print(f"\nScholarship : {scholarship_percent}%")
print(f"Amount      : ₹{scholarship_amount}")

print("\n===== END OF REPORT =====")
```

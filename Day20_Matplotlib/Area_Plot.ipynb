```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]

area1 = [2,4,6,8,10,12 ]

area2 = [3,4,6,2,7,3]

area3 = [1,3,3,5,3,5]

l = [ "area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"])
plt.title("Matplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```

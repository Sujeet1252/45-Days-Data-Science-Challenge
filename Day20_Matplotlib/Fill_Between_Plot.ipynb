```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6])
y = np.array([2,4,6,4,7,8])

plt.plot(x,y,color="r",label="Matplot")
plt.fill_between(x,y,color="y",where=(x>=2) & (x<=5),alpha=0.5)

plt.title("Matplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [20,40,26,38,10,12]


plt.boxplot(x,label = "Matplot",vert = True,widths = 0.5 , showmeans= True, sym = "g+",capprops = dict(color = "r"),whiskerprops = dict(color="r"))

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [20,40,26,38,10,12]

y= [10,20,30,40,50,60]

z = [x,y]
plt.boxplot(z,label = "Matplot",vert = True,widths = 0.5 , showmeans= True, sym = "g+",capprops = dict(color = "r"))

plt.legend()
plt.show()
```

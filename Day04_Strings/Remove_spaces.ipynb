```python
string = input("Enter a string: ")

space_count = string.count(" ")

result = string.replace(" ", "")

print("New String =", result)

print("Spaces Removed =", space_count)
```

```python
string = input("Enter a string: ")

result= ""

for char in string:
    if char != " ":
        result += char

print("String without spaces =", result)
```

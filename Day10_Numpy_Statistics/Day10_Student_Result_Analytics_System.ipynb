```python
import numpy as np 

print("=" * 50)
print("      STUDENT RESULT ANALYTICS SYSTEM")
print("=" * 50)

name = input("Enter Student Name :")

subjects = ["Mathematics",
            "Science",
            "Computer","Hindi","English"]

marks = []

for subject in subjects:
    mark = float(input(f"Enter Student Marks in {subject}:"))
    marks.append(mark)

marks = np.array(marks)

total_marks = np.sum(marks)

percentage = (total_marks / 500) * 100

average_marks = np.mean(marks)

highest_marks = np.max(marks)

lowest_marks = np.min(marks)

median_marks = np.median(marks)

variance_marks = np.var(marks)

std_deviation = np.std(marks)

if np.min(marks) < 33:
    status = "FAIL"
else:
    status = "PASS"

if percentage >= 90:
    grade = "A+"

elif percentage >= 80:
    grade = "A"

elif percentage >= 70:
    grade = "B"

elif percentage >= 60:
    grade = "C"

elif percentage >= 50:
    grade = "D"

else:
    grade = "F"

best_subject = subjects[np.argmax(marks)]
weak_subject = subjects[np.argmin(marks)]

print("\n")
print("=" * 50)
print("              RESULT REPORT")
print("=" * 50)

print(f"Student Name : {name}")

print("\nSubject Wise Marks")

for i in range(len(subjects)):
    print(f"{subjects[i]:20} : {marks[i]}")

print("\nStatistical Analysis")

print(f"Total Marks          : {total_marks}")
print(f"Percentage           : {percentage:.2f}%")
print(f"Average Marks        : {average_marks:.2f}")

print(f"Highest Marks        : {highest_marks}")
print(f"Lowest Marks         : {lowest_marks}")

print(f"Median               : {median_marks}")

print(f"Variance             : {variance_marks:.2f}")

print(f"Standard Deviation   : {std_deviation:.2f}")

print(f"\nBest Subject         : {best_subject}")
print(f"Weak Subject         : {weak_subject}")

print(f"\nGrade                : {grade}")
print(f"Result Status        : {status}")

print("=" * 50)
print("         END OF ANALYSIS REPORT")
```

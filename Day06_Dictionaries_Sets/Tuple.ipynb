```python
numbers = (78, 45, 99, 23, 67)

print("First Element :", numbers[0])

print("Last Element :", numbers[-1])

print("Length of Nummbers is: ", len(numbers))

print("Largest Number : ", max(numbers)) #O(n)

print("Smallest Number :", min(numbers))  #O(n)

```

```python
numbers = (78, 45, 99, 23, 67)

sorted_tuple = tuple(sorted(numbers))  #O(n log n)

print(sorted_tuple)

print("Smallest:", sorted_tuple[0])

print("Largest:", sorted_tuple[-1])
```

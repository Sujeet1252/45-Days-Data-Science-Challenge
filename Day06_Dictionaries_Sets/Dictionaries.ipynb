```python
#Creating Dictionaries

student = {
    "Name" : "Sujeet",
    "Course" : "MCA",
    "Age": 23

}

print(student)
print(student["Name"])
print(student["Age"])
```

```python
#Displaying all Keys

student = {
    "name": "Sujeet",
    "age": 23,
    "marks": 85
}

print(student.keys())

```

```python
#Displaying Values

student = {
    "name": "Sujeet",
    "age": 23,
    "marks": 85
}

print(student.values())
```

```python
#Updating Values

student = {
    "name": "Sujeet",
    "age": 23,
    "marks": 85
}
student["City"] = "Varanasi"

student["marks"] = 95

print(student)
```

```python
#Traversing Dictionary

for key in student:
    print(key)
```

```python
#Removing values

student = {
    "name": "Sujeet",
    "age": 23,
    "marks": 85
}

del student["age"]

print(student)
```

```python
#Print Key-Value Pairs Using Loop
student = {
    "name": "Sujeet",
    "age": 23,
    "marks": 85
}

for key, value in student.items():
    print(key, ":", value)
```

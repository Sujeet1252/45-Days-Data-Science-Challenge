```python
import pandas as pd

# Create DataFrame

data = {
    "Name": ["Aman", "Sujeet", "Ram", "Saurabh", "Pratham",
             "Sam", "Rohit", "Suresh", "Hritik"],

    "Marks": [85, 90, 95, 86, 80, 85, 56, 98, 85]
}

df = pd.DataFrame(data)

# Display Complete DataFrame
print("===== STUDENT DATA =====")
print(df)

# Head
print("\n===== FIRST 5 ROWS =====")
print(df.head())

# Tail
print("\n===== LAST 5 ROWS =====")
print(df.tail())

# Shape
print("\n===== SHAPE =====")
print(df.shape)

# Columns
print("\n===== COLUMN NAMES =====")
print(list(df.columns))

# Data Types
print("\n===== DATA TYPES =====")
print(df.dtypes)

# Information
print("\n===== DATAFRAME INFO =====")
df.info()

# Statistical Summary
print("\n===== STATISTICAL SUMMARY =====")
print(df.describe())
```

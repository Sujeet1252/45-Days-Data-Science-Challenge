```python
# ===========================
# Student Data Analyzer
# ===========================

import pandas as pd

data = {
    "Name" : ["Sujeet", "Sohit", "Sam", "Ram","Ravi"],
    "Marks" : [95,85,75,80,90]
}

df = pd.DataFrame(data)

print("Student Data")
print(df)

highest_marks = df["Marks"].max()
print("\n Highest Marks :", highest_marks)

lowest_marks = df["Marks"].min()
print("\n Lowest Marks :", lowest_marks)

average_marks = df["Marks"].mean()
print("\n Average Marks:", round(average_marks, 2))

topper = df.loc[df["Marks"].idxmax()]
print("\nTopper:")
print(topper)

lowest_scorer = df.loc[df["Marks"].idxmin()]
print("\nLowest Scorer:")
print(lowest_scorer)
```

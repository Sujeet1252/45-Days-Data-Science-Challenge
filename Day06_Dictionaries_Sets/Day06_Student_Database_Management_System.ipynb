```python
print("===== STUDENT DATABASE MANAGEMENT SYSTEM =====")

# Student Details
student = {}

student["name"] = input("Enter Student Name: ")
student["roll_no"] = input("Enter Roll Number: ")
student["course"] = input("Enter Course: ")

# Marks Input
marks = []

for i in range(1, 6):
    mark = float(input(f"Enter Marks of Subject {i}: "))
    marks.append(mark)

student["marks"] = marks

# Calculations
total = sum(marks)
average = total / len(marks)

highest = max(marks)
lowest = min(marks)

# Pass / Fail
if min(marks) < 33:
    status = "FAIL"
else:
    status = "PASS"

# Grade
if average >= 90:
    grade = "A+"
elif average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
else:
    grade = "D"

# Report
print("\n===== STUDENT REPORT =====")

print(f"Name       : {student['name']}")
print(f"Roll No    : {student['roll_no']}")
print(f"Course     : {student['course']}")

print(f"\nMarks      : {student['marks']}")

print(f"\nTotal      : {total}")
print(f"Average    : {average:.2f}")

print(f"Highest    : {highest}")
print(f"Lowest     : {lowest}")

print(f"\nGrade      : {grade}")
print(f"Status     : {status}")

print("\n===== END OF REPORT =====")
```

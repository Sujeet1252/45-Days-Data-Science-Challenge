```python
#1-D Array

import numpy as np

arr = np.array([10 , 20 , 30 , 40 , 50])

print(arr)
```

```python
#2-D Array

import numpy as np

arr = np.array([
    [1,5,6],
    [2,5,6]
])

print(arr)
```

```python
# Array Properties

arr = np.array([
        [1,2,3,4],
        [6,4,5,8]
])

print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
# Array Indexing

arr = np.array([10,20,30,40])

print(arr[0])
print(arr[-1])
```

```python
#Array Slicing

print(arr[1:3])
```

```python
# Array Operations

arr = np.array([1,2,3,4])

print(arr + 10)
print(arr * 2)
print(arr ** 2)
```

```python
# Array Reshape

arr = np.array([1,2,3,4,5,6])

print(arr.reshape(3,2))
print("\n")
print( arr.reshape(2,3))
```

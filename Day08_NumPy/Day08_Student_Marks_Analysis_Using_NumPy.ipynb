```python
import numpy as np

print("=" *35)
print(" STUDENT MARKS ANALYSIS SYSTEM ")
print("=" *35)


marks = []

for i in range(5):
    mark = float(input(f" Enter marks of Subject{i+1}: "))
    marks.append(mark)

marks = np.array(marks)

total_marks = np.sum(marks)
average_marks = np.mean(marks)
highest_marks = np.max(marks)
lowest_marks = np.min(marks)
median_marks = np.median(marks)
std_dev = np.std(marks)
variance = np.var(marks)

print("\n========== REPORT ==========")

print("Marks                :", marks)
print("Total Marks          :", total_marks)
print("Average Marks        :", round(average_marks, 2))
print("Highest Marks        :", highest_marks)
print("Lowest Marks         :", lowest_marks)
print("Median Marks         :", median_marks)
print("Standard Deviation   :", round(std_dev, 2))
print("Variance             :", round(variance, 2))

print("============================")
```

```python
import pandas as pd

import numpy as np

data = {
    "Name":["Sujeet","Rohan","Mohit","Nitin","Rohan"],
    "Age" : [23,34,25,25,np.nan],
    "Salary": [25000,30000,np.nan,20000,25000]
}

df= pd.DataFrame(data)

print(df)
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].fillna("27")
print(df)
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
print(df)
```

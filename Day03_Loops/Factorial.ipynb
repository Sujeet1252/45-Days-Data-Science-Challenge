```python
#using for loop

num= int(input("enter the number : "))

fact=1

for i in range(1,num+1):
    fact *= i 
print("Factorial is :", fact)
```

```python
#using while loop

num = int(input("Enter the number: "))

fact=1
i=1
while i<=num:
    fact *= i
    i+=1

print("factorial is : ", fact)
```

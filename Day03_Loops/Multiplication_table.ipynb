```python
#usin f-strings

n = int(input("Enter the value of n"))
for i in range(1,11):
    print(f"{n} * {i} = {n*i}")
    
```

```python
#without using f-string

n = int(input("Enter the value of n"))
for i in range(1,11):
    print(n, "*", i, "=", n*i)
```

```python
#using while loop

n = int(input("Enter the value of n"))
i=1
while i<=10:
    print(n, "*", i, "=", n * i)
    i +=1
```

```python
#using for loop

n = int(input("enter the number"))
a=0
b=1

print("Fibonacci series is:")

for i in range(n):
    print(a,  end=" ")
    c=a+b
    a=b
    b=c
```

```python
#using while loop

n = int(input("enter the number"))
a=0
b=1
count=0 

print("Fibonacci series is:")

while count<n:
    print(a,  end=" ")
    c=a+b
    a=b
    b=c
    count += 1
```

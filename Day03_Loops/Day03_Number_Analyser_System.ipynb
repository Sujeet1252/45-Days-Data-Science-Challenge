```python
# =========================
# Number Analyser System
# =========================

print("=== Number Analyser System ===")

num= int(input("Enter a Number :"))
if num==0:
    even_odd = "Restricted"
elif num%2== 0:
    even_odd = "Even"
else :
    even_odd = "Odd"


if num <= 1:
    prime = "Not possible"
else:
    for i in range(2,num):
        if  num %i ==0:
            prime = "No"
            break
    else:
        prime = "Yes"

factorial = 1
for i in range (1, num+1):
    factorial *= i

reverse = 0
original = num
while original>0:
    digit = original%10
    reverse = reverse * 10 + digit
    original  = original // 10


if reverse== num:
    palindrome="Yes"
else:
    palindrome="No"


original = num
sum_digit = 0
while original>0:
    digit = original%10
    sum_digit += digit
    original  = original // 10
    
print("=== Analysis Report ==\n")

print(f"Number  : {num}")
print(f"Even/odd   :{even_odd}")
print(f"Prime    : {prime}")
print(f"Factorial   :{factorial}")
print (f"Reverse   :{reverse}")
print(f"Palindrome   :{palindrome}")
print (f"Sum of Digits   :{sum_digit}")
        

    
    

```

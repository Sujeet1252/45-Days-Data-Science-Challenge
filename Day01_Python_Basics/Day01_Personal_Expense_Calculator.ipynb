```python
# ====================================
# PERSONAL EXPENSE CALCULATOR
# ====================================

print("===== PERSONAL EXPENSE CALCULATOR =====")

# User Input
name = input("Enter Your Name: ")

budget = float(input("Enter Monthly Budget (₹): "))

food = float(input("Food Expense (₹): "))
travel = float(input("Travel Expense (₹): "))
shopping = float(input("Shopping Expense (₹): "))
other = float(input("Other Expense (₹): "))

# Calculations
total_expense = food + travel + shopping + other
remaining_budget = budget - total_expense

# Status
if remaining_budget > 0:
    status = "Within Budget ✅"
elif remaining_budget == 0:
    status = "Budget Fully Utilized ⚠️"
else:
    status = "Budget Exceeded ❌"

# Report
print("\n===== EXPENSE REPORT =====")

print(f"Name              : {name}")
print(f"Monthly Budget    : ₹{budget}")

print(f"\nFood Expense      : ₹{food}")
print(f"Travel Expense    : ₹{travel}")
print(f"Shopping Expense  : ₹{shopping}")
print(f"Other Expense     : ₹{other}")

print(f"\nTotal Expense     : ₹{total_expense}")
print(f"Remaining Budget  : ₹{remaining_budget}")

print(f"\nStatus            : {status}")

print("\n===== END OF REPORT =====")
```

```python
# ==========================================
# STUDENT MANAGEMENT SYSTEM V1
# ==========================================

students = []
# Add Student
def add_student():

    name = input("Enter Name: ")
    roll = int(input("Enter Roll Number: "))

    marks = []

    for i in range(3):
        mark = float(input(f"Enter Marks of Subject {i+1}: "))
        marks.append(mark)

    student = {
        "name": name,
        "roll": roll,
        "marks": marks
    }

    students.append(student)

    print("\n Student Added Successfully")


# View Student by Roll Number
def view_student():

    roll = int(input("Enter Roll Number: "))

    for student in students:

        if student["roll"] == roll:

            average = sum(student["marks"]) / len(student["marks"])

            if average >= 90:
                grade = "A+"
            elif average >= 80:
                grade = "A"
            elif average >= 70:
                grade = "B"
            elif average >= 60:
                grade = "C"
            else:
                grade = "D"

            print("\n===== STUDENT DETAILS =====")
            print("Name     :", student["name"])
            print("Roll No  :", student["roll"])
            print("Marks    :", student["marks"])
            print("Average  :", round(average, 2))
            print("Grade    :", grade)

            return

    print(" Student Not Found")


# Search Student by Name
def search_student():

    name = input("Enter Student Name: ")

    for student in students:

        if student["name"].lower() == name.lower():

            average = sum(student["marks"]) / len(student["marks"])

            print("\n===== STUDENT FOUND =====")
            print("Name     :", student["name"])
            print("Roll No  :", student["roll"])
            print("Marks    :", student["marks"])
            print("Average  :", round(average, 2))

            return

    print(" Student Not Found")


# Show All Students
def show_all_students():

    if len(students) == 0:
        print("No Students Available")
        return

    print("\n===== ALL STUDENTS =====")

    for student in students:

        average = sum(student["marks"]) / len(student["marks"])

        print("--------------------------------")
        print("Name     :", student["name"])
        print("Roll No  :", student["roll"])
        print("Marks    :", student["marks"])
        print("Average  :", round(average, 2))


# Top Performer
def top_performer():

    if len(students) == 0:
        print(" No Students Available")
        return

    topper = students[0]
    highest_avg = sum(topper["marks"]) / len(topper["marks"])

    for student in students:

        avg = sum(student["marks"]) / len(student["marks"])

        if avg > highest_avg:
            highest_avg = avg
            topper = student

    print("\n TOP PERFORMER")
    print("Name     :", topper["name"])
    print("Roll No  :", topper["roll"])
    print("Average  :", round(highest_avg, 2))


# Main Menu
while True:

    print("\n===================================")
    print("     STUDENT MANAGEMENT SYSTEM")
    print("===================================")

    print("1. Add Student")
    print("2. View Student")
    print("3. Search Student")
    print("4. Show All Students")
    print("5. Top Performer")
    print("6. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_student()

    elif choice == "3":
        search_student()

    elif choice == "4":
        show_all_students()

    elif choice == "5":
        top_performer()

    elif choice == "6":
        print("\n Thank You!")
        print("Program Closed Successfully")
        break

    else:
        print("Invalid Choice")
```

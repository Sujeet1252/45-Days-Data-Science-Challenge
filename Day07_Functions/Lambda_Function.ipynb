```python
add = lambda a, b : a + b

print(add(10,20))
```

```python
#Without lambda

def square(x):
    return x**2
result = square(5)
print(result)
```

```python
#with lambda

square = lambda x:x**2

print(square(5))
```

```python
# Add two number

add = lambda a ,b : a + b

print(add(5,7))
```

```python
# Larger nummber

largest = lambda a, b : a if a> b else b

print(largest(5,7))
```

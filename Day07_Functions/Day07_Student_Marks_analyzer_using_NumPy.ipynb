```python
import numpy as np

# Student Marks Data
marks = np.array([78, 90, 67, 88, 95, 72])

# Analysis
total_marks = np.sum(marks)
average_marks = np.mean(marks)
highest_marks = np.max(marks)
lowest_marks = np.min(marks)

passed_students = np.sum(marks >= 40)
failed_students = np.sum(marks < 40)

above_average_students = np.sum(marks > average_marks)

# Report
print("=" * 40)
print("      STUDENT MARKS ANALYZER")
print("=" * 40)

print(f"Total Students         : {len(marks)}")
print(f"Total Marks            : {total_marks}")
print(f"Average Marks          : {average_marks:.2f}")
print(f"Highest Marks          : {highest_marks}")
print(f"Lowest Marks           : {lowest_marks}")
print(f"Passed Students        : {passed_students}")
print(f"Failed Students        : {failed_students}")
print(f"Students Above Average : {above_average_students}")

print("=" * 40)
```

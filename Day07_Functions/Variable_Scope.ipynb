```python
# Local Variable

def greet():
    name = " sujeet"  #local variable

    print(name)   # inside a function

greet()
```

```python
def greet():
    name = " sujeet" 

greet()

print(name)   # outside a function
```

```python
# Global Variable

name = "Sujeet" # global variable

def greet():
    print(name)

greet()  
print(name)
```

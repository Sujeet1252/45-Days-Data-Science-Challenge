```python
#Find Sum
import numpy as np

arr = np.array([10,50,60,20,30,40,50,40])


print(np.sum(arr))
```

```python
#Find Mean

print(np.mean(arr))
```

```python
#Find Maximum

print(np.max(arr))
```

```python
#Find Minimum

print(np.min(arr))
```

```python
#Multiply all element by two

print(arr *2 )
```

```python
# Add 100 in each element
arr= np.array(arr)
print(arr + 100)
```

```python
def add(a,b):
    return a + b

result = add(12,52)

print(result)
```

```python
#Find Larger Number

def largest(a,b):
    if a>b:
        return a
    else:
        return b
print("Largest number is :", largest(10,30))
```

```python
# Even or Odd

def number(num):
    if num %2 == 0:
        return "Even"
    else:
        return "Odd"

print("Number is : ", number(25))
```

```python
def square(x):
    return x**2
result = square(5)
print(result)
```

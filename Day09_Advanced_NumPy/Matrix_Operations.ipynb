```python
import numpy as np

arr1 = np.array([[4,7],[8,9]])
print("First matrix :", arr1)

arr2 = np.array([[5,6],[9,3]])
print("Second matrix :", arr2)


print("\nSum of both matrix : ")
print(arr1+arr2)

print("\nMatrix Multiplication :")
print(np.dot(arr1,arr2))

print("\nTranspose of first matrix is :")
print(arr1.T)
```

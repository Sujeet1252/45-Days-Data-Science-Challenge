```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = sns.load_dataset("titanic")
dataset.head(10)
```

```python
dataset["age"].skew()
```

```python
#Left Skewness
sns.histplot(x="age",data=dataset)
plt.show()
```

```python
data = np.random.normal(0,100,100)
```

```python
df = pd.DataFrame({"x":data})
df["x"].skew()
```

```python
#Right Skewness

sns.histplot(x="x",data=df)
plt.show()
```

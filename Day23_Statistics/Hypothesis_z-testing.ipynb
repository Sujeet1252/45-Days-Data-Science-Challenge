```python
import scipy.stats as st
import numpy as np
```

```python
x = 90
u = 82
std = 20
n = 81
ap = 0.05
```

```python
z_calc  = (x-u)/(std/np.sqrt(n))
z_calc
```

```python
z_table = st.norm.ppf(1-ap)
z_table
```

```python
if z_table < z_calc :
    print(" Ha is right")
else :
    print(" Ha is not")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [np.random.randint(10,100) for i in range(1000)]
table = pd.DataFrame({"data": data})
table
```

```python
plt.figure(figsize = (3,2))
sns.kdeplot(x="data",data = table)
plt.show()
```
